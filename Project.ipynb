```python
zipcodes = list(set(data["regionidzip"]))
zipcode_data = {}
fields = ["value", "bedrooms", "lotsize", "squarefootage", "count"]
for zipcode in zipcodes:
    temp_dict = {}
    for field in fields:
        temp_dict[field] = 0
    zipcode_data[zipcode] = temp_dict
```

```python
for index, row in data.iterrows():
    zipcode_data[row["regionidzip"]]["value"] += row["taxvaluedollarcnt"]
    zipcode_data[row["regionidzip"]]["bedrooms"] += row["bedroomcnt"]
    zipcode_data[row["regionidzip"]]["lotsize"] += row["lotsizesquarefeet"]
    zipcode_data[row["regionidzip"]]["squarefootage"] += row["calculatedfinishedsquarefeet"]
    zipcode_data[row["regionidzip"]]["count"] += 1

zipcode_data
```

```python
for zipcode in zipcodes:
    count = zipcode_data[zipcode]["count"]
    zipcode_data[zipcode]["value"] = zipcode_data[zipcode]["value"] / count
    zipcode_data[zipcode]["bedrooms"] = zipcode_data[zipcode]["bedrooms"] / count
    zipcode_data[zipcode]["lotsize"] = zipcode_data[zipcode]["lotsize"] / count
    zipcode_data[zipcode]["squarefootage"] = zipcode_data[zipcode]["squarefootage"] / count
```

```python
X = df[:,:-1]
y = df[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0, shuffle=True)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

predicted_y = regr.predict(X_test)
acc = regr.score(X_test,y_test)
y_diff = predicted_y - y_test
y_diff
```

```python
results = []
for i in range(len(y_diff)):
    results.append((y_diff[i] / X_test[i][0]) * 100)
    
bucket_results = {}
bucket_results["<1"] = 0
bucket_results["1-2.5"] = 0
bucket_results["2.5-5"] = 0
bucket_results["5-10"] = 0
bucket_results[">10"] = 0
for result in results:
    if(abs(result) < 1):
        bucket_results["<1"] += 1
    elif(abs(result) < 2.5):
        bucket_results["1-2.5"] += 1
    elif(abs(result) < 5):
        bucket_results["2.5-5"] += 1
    elif(abs(result) < 10):
        bucket_results["5-10"] += 1
    else:
        bucket_results[">10"] += 1
        
bucket_results
```

```python
y_train_log = []
y_test_log = []

for i in range(len(y_train)):
    num = abs(y_train[i] / X_train[i][0]) * 100
    result = -1
    if(num < 5):
        y_train_log.append(0)
    elif(num < 10):
        y_train_log.append(1)
    elif(num < 15):
        y_train_log.append(2)
    else:
        y_train_log.append(3)
    
for i in range(len(y_test)):
    num = abs(y_test[i] / X_test[i][0]) * 100
    result = -1
    if(num < 5):
        y_test_log.append(0)
    elif(num < 10):
        y_test_log.append(1)
    elif(num < 15):
        y_test_log.append(2)
    else:
        y_test_log.append(3)
        
y_train_log
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression

logreg = linear_model.LogisticRegression(multi_class='multinomial', solver='lbfgs')
logreg.fit(X_train, y_train_log)

predicted_y_log = logreg.predict(X_test)
acc_log = logreg.score(X_test,y_test_log)

predicted_y_log
```

```python
from sklearn.metrics import confusion_matrix
Cmat = confusion_matrix(y_test_log, predicted_y_log)
acc = logreg.score(X_test,y_test_log)
print("Confusion Matrix: ")
print(Cmat)
print("Accuracy: " + str(acc*100) + '%')
```

```python
import seaborn as sns

ax = sns.heatmap(Cmat/np.sum(Cmat), annot=True, 
            fmt='.2%', cmap='Blues')

ax.set_title('Logistic Regression for Classifying Price Change Within 5%\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
